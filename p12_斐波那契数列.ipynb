```python
def fib1(n):
    if n < 1:
        return -1
    if n == 1 or n == 2:
        return 1
    return fib1(n-1) + fib1(n-2)
```

```python
@utils.calc_runing_time
def run_fib1(n):
    return fib1(n)
```

```python
run_fib1(40)
```

```python
@utils.calc_runing_time
def fib2(n):
    if n < 1:
        return -1
    fib_list = [1,1]
    for i in range(3, n+1):
        fib_list.append(fib_list[-1] + fib_list[-2])
    return fib_list
```

```python
fib2(40)[-1]
```

```python
@utils.calc_runing_time
def fib3(n):
    if n < 1:
        return -1
    s1 = 1
    s2 = 1
    for i in range(3,n+1):
        s2 = s1 + s2
        s1 = s2 - s1
    return s2
```

```python
fib3(40)
```
