# 递归算法

```python
import utils
```

```python
def fib1(n):
    if n < 1:
        return -1
    if n == 1 or n == 2:
        return 1
    return fib1(n-1) + fib1(n-2)
```

```python
@utils.calc_runing_time
def run_fib1(n):
    return fib1(n)
```

```python
run_fib1(40)
```

# 循环
时间复杂度为$O(n)$, 空间复杂度为$O(n)$

```python
@utils.calc_runing_time
def fib2(n):
    if n < 1:
        return -1
    fib_list = [1,1]
    for i in range(3, n+1):
        fib_list.append(fib_list[-1] + fib_list[-2])
    return fib_list
```

```python
fib2(40)[-1]
```

算法使用列表把所有的中间结果记录，其实算法本身不需要记录所有的中间结果。\
这样算法的时间复杂度不变，空间复杂度降低为$O(1)$

```python
@utils.calc_runing_time
def fib3(n):
    if n < 1:
        return -1
    s1 = 1
    s2 = 1
    for i in range(3,n+1):
        s2 = s1 + s2
        s1 = s2 - s1
    return s2
```

```python
fib3(40)
```
